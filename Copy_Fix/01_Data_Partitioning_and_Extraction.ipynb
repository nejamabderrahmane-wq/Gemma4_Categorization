```python
# Dataset Splitting & Categorization Pipeline

**Objective:**
This notebook processes a centralized dataset containing e-commerce transaction records. It reads a master `.xlsx` file and partitions the data into two distinct datasets based on the platform of origin (`service_name`):

1. **Categorized.xlsx**: Contains all records originating from 'mercari'.
2. **Non_Categorized.xlsx**: Contains all remaining records from other platforms.

This split isolates data that already possesses a complete taxonomy structure from data that requires downstream AI-driven categorization.
```

```python
# Install required dependency if not already present
!pip install openpyxl -q

import os
import pandas as pd

# --- Configuration & File Paths ---
# INSTRUCTION: Update these placeholder paths to match your local or cloud environment.
INPUT_FILE_PATH = 'path/to/your/input_data/Rename.xlsx'
OUTPUT_DIR = 'path/to/your/output_directory/Split_Files'

# Target variable for classification splitting
TARGET_SERVICE = 'mercari'

# Ensure the output directory exists securely
os.makedirs(OUTPUT_DIR, exist_ok=True)

print("Dependencies loaded and configuration set.")
```

```python
# --- Data Ingestion ---
print(f"Loading data from: {INPUT_FILE_PATH}...")
try:
    df = pd.read_excel(INPUT_FILE_PATH)
    
    # Display basic diagnostic information
    unique_services = df['service_name'].dropna().unique()
    print(f"Total rows loaded: {len(df)}")
    print(f"Total unique services detected: {len(unique_services)}")
    print(f"Services found: {list(unique_services)}")
    
except FileNotFoundError:
    print(f"Error: The file at {INPUT_FILE_PATH} was not found. Please verify the path.")
```

```python
# --- Data Partitioning ---
# Utilizing pandas boolean indexing for highly optimized data splitting

# 1. Filter the dataset based on the target condition
categorized_df = df[df['service_name'].astype(str).str.lower() == TARGET_SERVICE.lower()]
non_categorized_df = df[df['service_name'].astype(str).str.lower() != TARGET_SERVICE.lower()]

# 2. Define standard output file paths
categorized_path = os.path.join(OUTPUT_DIR, 'Categorized.xlsx')
non_categorized_path = os.path.join(OUTPUT_DIR, 'Non_Categorized.xlsx')

# 3. Export to Excel
print("Exporting datasets...")
categorized_df.to_excel(categorized_path, index=False)
non_categorized_df.to_excel(non_categorized_path, index=False)

# 4. Summary Validation
print("\n--- Pipeline Execution Summary ---")
print(f"Categorized records ('{TARGET_SERVICE}'): {len(categorized_df)} rows -> Saved to {categorized_path}")
print(f"Non-Categorized records (Others): {len(non_categorized_df)} rows -> Saved to {non_categorized_path}")
print("Process completed successfully.")
```
